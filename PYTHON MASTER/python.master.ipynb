# Lecture 2 - Libraries and Datasets

```python
# math library (typically covered with numpy)
import math
print(math.log2(8))
```

```python
# statistics library (typically covered with numpy)
import statistics
print(statistics.mean([1, 7, 3, 5, 9, 1, 12]))
print(statistics.median([1, 7, 3, 5, 9, 1, 12]))
print(statistics.geometric_mean([1, 7, 3, 5, 9, 1, 12]))
```

```python
# give the libraries / modules nicknames
import matplotlib.pyplot as plt # this library is important for data visualization
import pandas as pd # this library is important for working with data
```

```python
# reading and working with datasets
import pandas as pd
carfeatures = pd.read_csv("data copy/features.csv") # reads in features dataset
display(carfeatures) # Entering the name of a dataframe produces an output with some rows
display(carfeatures['cylinders']) # extract cylinders variable
help (pd.crosstab) # the "help" keyword can be used to learn more about a command
```

```python
# crosstab counts how many rows fall into categories
# "index" is the category
# "columns" is a custom title
table = pd.crosstab(index = carfeatures['cylinders'],columns = "count")
display(table)

# we can also crosstabulate between 2 columns of carfeatures
table_2 = pd.crosstab(index = carfeatures['cylinders'],columns = carfeatures['mpg'])
display(table_2)
```

```python
# "describe" computes the count, mean, std, min, 25% quantile, 50%, 75%, max
# automatically excludes variables with text values
# otherwise includes all numeric variables

display(carfeatures.describe())
display(round(carfeatures.describe(), 2)) # rounds to two digits
display(round(carfeatures.describe(percentiles=[.33,.66]), 2)) # changes code to print 33 and 66 percentile instead of automatic 25 and 75
```

```python
# scatterplot
plt.scatter(x = carfeatures['weight'], y = carfeatures['mpg'])
plt.show()
```

```python
# we can use the built in type functions to check an object's type
print(type(3)) # integer
print(type(3.25)) # float
print(type("Hello World!")) # string
print(type(print)) # built-in function
```

```python
# the = in python is NOT AN EQUAL SIGN. IT IS CALLED AN ASSIGNMENT OPERATOR
x = 3 # takes the integer 3 and assigns it to a new variable x
y = 3.25 # takes the float 3.25 and assigns it to a new variable y
# everything to the right of "=" happens first
print(x)
print(y)
```

```python
# we can check our variables by pressing jupyter variables as the top
record_price = 32.99
fulton_tax_rate = 0.089
dekalb_tax_rate = 0.08

fulton_cost = record_price + record_price * fulton_tax_rate
fulton_cost = round(fulton_cost, 2) # round to 2 decimal places
print(fulton_cost)

dekalb_cost = record_price + record_price * dekalb_tax_rate
dekalb_cost = round(dekalb_cost, 2) # round to 2 decimal places
print(dekalb_cost)


price_difference = fulton_cost - dekalb_cost
print(round(price_difference, 2)) # need to re round the final result as well
```

# Lecture 5 - Tuples, Strings

```python
list_of_numbers = [5, 3, 2, 0]
tuple_of_numbers = (5, 3, 2, 0)

print(list_of_numbers) # lists are changeable
print(tuple_of_numbers) # tuples are not changeable
print()
print(type(list_of_numbers))
print(type(tuple_of_numbers)) # can also make tuples without explicitly including parenthesis
```

```python
# we can also use strings, which are sequences of characters
my_name = "Natasha Strauss"

print(type(my_name)) 
print()
print(my_name[0]) # Character zero of string (a capital "N") 
print(my_name[4]) # Character four of string (lowercase "s")
print(my_name[5]) # Character five of string (lowercase "h")
print(my_name[7]) # Character six of string (space)
print(my_name[10]) # Character ten of string (lowercase "r")

# print(my_name[67]) there is no Character 67 so it would cause an error
# python starts all counts from zero
# strings are not changeable
```
