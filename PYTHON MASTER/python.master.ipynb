```python
# reading and working with datasets
import pandas as pd
carfeatures = pd.read_csv("data copy/features.csv") # reads in features dataset
display(carfeatures) # Entering the name of a dataframe produces an output with some rows
display(carfeatures['cylinders']) # extract cylinders variable
help (pd.crosstab) # the "help" keyword can be used to learn more about a command
```

```python
# crosstab counts how many rows fall into categories
# "index" is the category
# "columns" is a custom title
table = pd.crosstab(index = carfeatures['cylinders'],columns = "count")
display(table)

# we can also crosstabulate between 2 columns of carfeatures
table_2 = pd.crosstab(index = carfeatures['cylinders'],columns = carfeatures['mpg'])
display(table_2)
```

```python
# "describe" computes the count, mean, std, min, 25% quantile, 50%, 75%, max
# automatically excludes variables with text values
# otherwise includes all numeric variables

display(carfeatures.describe())
display(round(carfeatures.describe(), 2)) # rounds to two digits
display(round(carfeatures.describe(percentiles=[.33,.66]), 2)) # changes code to print 33 and 66 percentile instead of automatic 25 and 75
```

```python
# import the library "np" is the standard nickname
import numpy as np

# the most common way to create a numpy array is to convert a list into a numpy array

list_of_numbers = [5.6, 2.9, 9.7, -10, 0, 3]

array_of_numbers = np.array(list_of_numbers)

print(type(array_of_numbers))
print(array_of_numbers)
display(array_of_numbers)

# usually, we would create the numpy array in a single line, without defining a separate list first

array_of_numbers2 = np.array([5.6, 2.9, 9.7, -10, 0, 3])


# you can access elements of arrays in the same way as lists:
print(array_of_numbers2[0])
print(array_of_numbers2[4])
```

```python
# numpy arrays are nice because you can do numerical operations on them
array_1 = np.array([2, 3, 5, 11])
array_2 = np.array([1, -4, 2, 0])

# the "+" operation adds the elements of the two arrays together
print(array_1 + array_2)

# can also do operations with a constant
a = np.array([3, 11, -2.2])
print(a + 2)
```

```python
my_var = 2

print(np.log(my_var)) #ln(x)
print(np.exp(my_var)) #e^x
print(np.sin(my_var)) # sin(x) in radians
print(np.cos(my_var)) # cos(x) in radians
print(np.sqrt(my_var)) # square root of x

print(np.pi) # pi = circle's circumference/circle's diameter

print(np.e) # Euler's number = limit of (1 + 1/n)^n as n goes to infinity

print(np.exp(1)) # another way to calculate Euler's number

# or you can just use a 2
print(np.log(2))
```

```python
# plotting with numpy
# Create list of equally spaced points from -1.5 to 1.5
# We'll choose points 0.25 apart
x_vals = [-1.5, -1.25, -1, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75, 1, 1.25, 1.5]

# Convert to numpy array
x_vals = np.array(x_vals)

# Create new numpy array with the function evaluated at the same points
y_vals = x_vals**3 - 2*x_vals + 0.5

# Use plt.scatter to create scatter plot
plt.scatter(x = x_vals, y = y_vals)
plt.show()

# convert list to numpy array right away
t_vals = np.array( [0, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5] )

# Evaluate function and create new array 
g_of_t = np.exp(t_vals**2) - np.sqrt(t_vals) + 1

# Use plt.scatter to create scatter plot
plt.scatter(x = t_vals, y = g_of_t)
plt.show()
```

```python
# Boolean Objects (true or false)
print(type(True))
print(type(False))

print()

# Adding quotes results in string objects
print(type("True"))
print(type('False'))
```

```python
# comparison operators
print(5>3)
print(6<=4)
print (9<3)
print(2>=7)
print(5*3 == 15)
print(3 - 2 != 7) #not equal
```

```python
# booleans with strings
x = "Hello World!"
y = "Hello World !"

print(x == "Hello World!")
print(y == "Hello World!")

print(x != y)

print('ab' == 'ba')

list_1 = ["blue", "red", "green"]
list_2 = ["blue", "red", "green"]
list_3 = ["red", "green", "blue"]
list_4 = ["red", "green", "blue", "violet"]

print(list_1 == list_2)
print(list_1 == list_3) # False: different ordering
print(list_3 == list_4) # False: different number of elements
print(list_3 != list_4) # True
```

```python
# when comparing numpy arrays, a seperate boolean expression is returned for each element
vec_a = np.array([1,2,3])
vec_b = np.array([1,2,4])

print(vec_a == vec_b)

vec_c = np.array([10, 5, 3])
vec_d = np.array([20, 10, 8])
print(vec_c >= vec_d)
```

```python
# The first way to use the "in" command is to check whether a word is contained in a sentence. This can be useful if you're trying to search for patterns

sentence = "The Federal Reserve makes forecasts about many economic outcomes"


keyword = "economic"
print(keyword in sentence)

keyword = "bank"
print(keyword in sentence)

search_word = "Federal" 
print(search_word in sentence)

important_word = "federal"
print(important_word in sentence)

print("F" in sentence)

# Use the keyword "in" to see if a number belongs to a NumPy Array
# (Also works for lists)

s = np.array([1.0, 3.0, -17, 12])
print(1 in s)

t = [0, 22, "PAIS 557"]

print(2 in t)
# In the previous cell, "1 in s" evaluated to True because the integer 1 and the floating point number 1.0 are regarded as equal
```

```python
# logical operators: not, and, or
age  = 22

# True or false? This person can vote in the U.S.

print(not age < 18)

# Sometimes introducing an intermediate variable can make things more clear

age = 22 

statement = age < 18
print(statement) 

print(not statement)

# Sometimes introducing an intermediate variable can make things more clear

age = 22 

statement = age < 18
print(statement) 

print(not statement)

price = 24.5

list_prices = [32.5, 13, 0.99, 15.96]

# True or false? The price is greater than 10 dollars and is one of the elements of list_prices

print(price > 10 and price in list_prices)

# or checks or

statement_1 = price > 10 
statement_2 = price in list_prices

print(statement_1)
print(statement_2)
print(statement_1 or statement_2)

# can seperate with parentheses for readability
x = 13
y = -5
z = False

P = (x > y) and (not z) and (x*y < 0)
print(P)

Q = (x == 5) or (y > 10) or (x + y > 0)
print(Q)
```

```python
# create empty lists
good_movies = [] 
ok_movies = []
bad_movies = []

cut_off_1 = 7.0 # minimum rating for "good movie"
cut_off_2 = 6.0 # minimum rating for "ok movie"


title_1 = "The Umbrellas of Cherbourg"
imdb_rating_1 = 7.8 

title_2 = "M3gan" 
imdb_rating_2 = 6.3 

title_3 = "Psycho" 
imdb_rating_3 = 8.5 

title_4 = "Melania"
imdb_rating_4 = 1.3

# Divide movies between the three lists
# print a warning if we encounter a bad movie

if imdb_rating_1 >= cut_off_1:
    good_movies = good_movies + [title_1] # add movie to good_movies
elif imdb_rating_1 >= cut_off_2:
    ok_movies = ok_movies + [title_1] # add movie to ok_movies
else:
    bad_movies = bad_movies + [title_1] # add movie to bad_movies
    print("Warning! Bad movie alert:", title_1) # ALSO: print warning message

if imdb_rating_2 >= cut_off_1:
    good_movies = good_movies + [title_2]  
elif imdb_rating_2 >= cut_off_2:
    ok_movies = ok_movies + [title_2]
else:
    bad_movies = bad_movies + [title_2]
    print("Warning! Bad movie alert:", title_2)


if imdb_rating_3 >= cut_off_1:
    good_movies = good_movies + [title_3] 
elif imdb_rating_3 >= cut_off_2:
    ok_movies = ok_movies + [title_3]
else:
    bad_movies = bad_movies + [title_3]
    print("Warning! Bad movie alert:", title_3)


if imdb_rating_4 >= cut_off_1:
    good_movies = good_movies + [title_4] 
elif imdb_rating_4 >= cut_off_2:
    ok_movies = ok_movies + [title_4]
else:
    bad_movies = bad_movies + [title_4]
    print("Warning! Bad movie alert:", title_4)


print(good_movies)
print(ok_movies)
print(bad_movies)
```

```python
# create lists
numerical_list = [5.3, 25, -0.1, -5, 11, 3, 1.2]
string_list = ["Frank", "Zev", "Peter", "Abigail", "Ali", "Eleanor"]

# sort them
numerical_list.sort()
string_list.sort()

# print out sorted versions
print(numerical_list)
print(string_list)
```

```python
# copy a list
list_of_floats = [2.3, -1.0, 5.1, 9.9]

copy_of_list = list_of_floats.copy()

list_of_floats[0] = "Nothing"

print(copy_of_list)

# reverse the order of a list
colors = ["blue", "green", "red", "orange", "pink", "purple"]

colors.reverse()

print(colors)

# count the number of times an item appears in the list
fav_colors = ["blue", "pink", "black", "pink", "red", "blue", "green", "black", "pink"]

pink_count = fav_colors.count("pink")

print(pink_count) # prints "3", the number of times "pink" appears in the list
```

```python
# summary statistics
c_vec = np.array([4.2, 1.2, 11.45, -2, 0, 9, 3.3, 2*np.pi])


print(np.mean(c_vec)) # mean
print(np.std(c_vec)) # standard deviation
print(np.median(c_vec)) # median
print(np.max(c_vec)) # maximum
print(np.min(c_vec)) # minimum

# another method

print(c_vec.mean())
print(c_vec.std())

# arrays don't have a .median() method for some reason...

print(c_vec.max())
print(c_vec.min())
```

```python
# Generate a single sample from a standard Normal Distribution
# "Standard Normal" means the mean is zero, and the standard deviation is one.

normal_sample = np.random.normal()
print(normal_sample)

# generate 5 samples from a normal distribution with standard deviation = 2.4, and mean = -1
five_samples = np.random.normal(loc = -1, scale = 2.4, size = 5)
print(five_samples)

# run this cell 3 times, numbers will always be the same
np.random.seed(0) # set seed for reproduce
print(np.random.normal(loc = 0.5, scale = 3, size = 4))
```

```python
import matplotlib.pyplot as plt

np.random.seed(1989)


sample_size = 1000

standard_normal_samples = np.random.normal(size = sample_size)

plt.hist(x = standard_normal_samples)
plt.xlabel("Values drawn from Distribution")
plt.ylabel("Frequency")
plt.title("Histogram of standard normal variable (1,000 samples)")
plt.show()
```
